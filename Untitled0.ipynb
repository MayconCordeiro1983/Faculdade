```python
# ========================================
# ETAPA 1 - LIMPEZA E ORGANIZAÇÃO DOS DADOS
# ========================================

import pandas as pd
import numpy as np
import re

# 1️⃣ Leitura do arquivo (ajuste o caminho se necessário)
# Envie o arquivo "importacoes-exportacoes-derivados-2000-2025.csv" para o Colab e depois rode:
file_path = "importacoes_exportacoes_limpos_etapa1_corrigido.csv"

# Leitura com separador ';'
df_raw = pd.read_csv(file_path, sep=',', encoding='utf-8', dtype=str)
print("Registros originais:", len(df_raw))

# ----------------------------------------
# 2️⃣ Padronização de nomes de colunas
# ----------------------------------------
df_raw.columns = [c.strip().upper() for c in df_raw.columns]

rename_map = {}
for c in df_raw.columns:
    if 'ANO' in c:
        rename_map[c] = 'ANO'
    elif 'MÊS' in c or 'MES' in c:
        rename_map[c] = 'MES'
    elif 'PRODUT' in c:
        rename_map[c] = 'PRODUTO'
    elif 'OPERA' in c:
        rename_map[c] = 'OPERACAO_COMERCIAL'
    elif 'DISP' in c or 'RECEITA' in c or 'US$' in c or 'FOB' in c:
        rename_map[c] = 'DISPENDIO_USD_FOB'
    elif 'IMPORTADO' in c or 'EXPORTADO' in c or 'IMPORTADO / EXPORTADO' in c:
        rename_map[c] = 'VOLUME_M3'

df = df_raw.rename(columns=rename_map)

# ----------------------------------------
# 3️⃣ Conversões numéricas e limpeza
# ----------------------------------------
def to_numeric_safe(x):
    try:
        if pd.isna(x): return np.nan
        if isinstance(x, str):
            x2 = x.replace('.', '').replace(',', '.')
            x2 = re.sub(r'[^\d\.-]', '', x2)
            if x2 in ('', '.', '-'): return np.nan
            return float(x2)
        return float(x)
    except:
        return np.nan

df['VOLUME_M3'] = df['VOLUME_M3'].apply(to_numeric_safe)
df['DISPENDIO_USD_FOB'] = df['DISPENDIO_USD_FOB'].apply(to_numeric_safe)

# Converte ANO
df['ANO'] = pd.to_numeric(df['ANO'], errors='coerce')
df = df.dropna(subset=['ANO', 'MES']).copy()
df['ANO'] = df['ANO'].astype(int)

# Remove duplicatas
df = df.drop_duplicates()

# Trata valores zero ou negativos
df.loc[df['VOLUME_M3'] <= 0, 'VOLUME_M3'] = np.nan
df.loc[df['DISPENDIO_USD_FOB'] <= 0, 'DISPENDIO_USD_FOB'] = np.nan

print("Registros após limpeza:", len(df))
print("Registros com volume válido:", df['VOLUME_M3'].notna().sum())
print("Registros com dispêndio válido:", df['DISPENDIO_USD_FOB'].notna().sum())

# ----------------------------------------
# 4️⃣ Período temporal coberto
# ----------------------------------------
ano_min = int(df['ANO'].min())
ano_max = int(df['ANO'].max())
print(f"Período temporal: {ano_min} a {ano_max}")

# ----------------------------------------
# 5️⃣ Outliers (média ± 3 * desvio padrão)
# ----------------------------------------
def detectar_outliers(col):
    media = df[col].mean(skipna=True)
    desvio = df[col].std(skipna=True)
    limite_inferior = media - 3 * desvio
    limite_superior = media + 3 * desvio
    outliers = df[(df[col] < limite_inferior) | (df[col] > limite_superior)][col]
    return media, desvio, len(outliers), limite_inferior, limite_superior

# Volume
media_vol, desvio_vol, num_out_vol, lim_inf_vol, lim_sup_vol = detectar_outliers('VOLUME_M3')
# Dispêndio
media_disp, desvio_disp, num_out_disp, lim_inf_disp, lim_sup_disp = detectar_outliers('DISPENDIO_USD_FOB')

print("\n=== Outliers ===")
print(f"Volume (m³): média={media_vol:.2f}, desvio={desvio_vol:.2f}, outliers={num_out_vol}")
print(f"Dispêndio (US$ FOB): média={media_disp:.2f}, desvio={desvio_disp:.2f}, outliers={num_out_disp}")

# ----------------------------------------
# 6️⃣ Exemplo de visualização
# ----------------------------------------
print("\nAmostra dos dados limpos:")
df.head()
```

```python
# =====================================================
# ETAPA 2 - ANÁLISE ESTATÍSTICA DESCRITIVA COMPLETA
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------------------------
# 1️⃣ Leitura dos dados já limpos (Etapa 1)
# -----------------------------------------------------
file_path = "importacoes_exportacoes_limpos_etapa1_corrigido.csv"
df = pd.read_csv(file_path)

print(f"Registros carregados: {len(df)}")
print(f"Colunas disponíveis: {list(df.columns)}\n")

# =====================================================
# 2️⃣ Quartis, IQR e Meses do Quartil Superior
# =====================================================
q1 = df['VOLUME_M3'].quantile(0.25)
q2 = df['VOLUME_M3'].quantile(0.5)
q3 = df['VOLUME_M3'].quantile(0.75)
iqr = q3 - q1

print("=== Quartis e Intervalo Interquartil (IQR) ===")
print(f"Q1 = {q1:,.2f}")
print(f"Q2 (Mediana) = {q2:,.2f}")
print(f"Q3 = {q3:,.2f}")
print(f"IQR = {iqr:,.2f}\n")

# Meses no quartil superior
quartil_superior = df[df['VOLUME_M3'] > q3]
meses_quartil_superior = (
    quartil_superior['MES']
    .value_counts()
    .sort_values(ascending=False)
)

print("=== 1️⃣ Meses que representam o quartil superior ===")
print(f"Limite do quartil superior (Q3): {q3:,.2f} m³\n")
print("Meses mais frequentes entre os 25% maiores volumes:")
print(meses_quartil_superior)
print("\nInterpretação:")
print("Os meses acima concentram os 25% maiores volumes, indicando períodos de maior movimentação comercial.\n")

# =====================================================
# 3️⃣ Outliers extremos (Critério 1,5 × IQR)
# =====================================================
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr
outliers_iqr = df[(df['VOLUME_M3'] < lim_inf) | (df['VOLUME_M3'] > lim_sup)]

print("=== 2️⃣ Evidência de outliers extremos (1,5 × IQR) ===")
print(f"Limite inferior = {lim_inf:,.2f}")
print(f"Limite superior = {lim_sup:,.2f}")
print(f"Foram encontrados {len(outliers_iqr)} registros fora dos limites (1,5 × IQR).")

if len(outliers_iqr) > 0:
    print("✅ Há evidência de outliers extremos na série, representando operações fora do padrão.")
else:
    print("❎ Não há evidência de outliers extremos.\n")

# Boxplot visual dos outliers
plt.figure(figsize=(8,4))
sns.boxplot(y=df['VOLUME_M3'], color='skyblue')
plt.title("Boxplot - Detecção de Outliers (Volume m³)")
plt.ylabel("Volume (m³)")
plt.tight_layout()
plt.show()

# =====================================================
# 4️⃣ Sazonalidade - Análise de Volume Médio Mensal
# =====================================================
df_mes = df.groupby('MES', as_index=False)['VOLUME_M3'].mean()

# Ordem cronológica dos meses
ordem_meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho',
               'Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']
df_mes['MES'] = pd.Categorical(df_mes['MES'], categories=ordem_meses, ordered=True)
df_mes = df_mes.sort_values('MES')

plt.figure(figsize=(10,4))
sns.lineplot(data=df_mes, x='MES', y='VOLUME_M3', marker='o', color='teal', linewidth=2)
plt.title('Volume Médio Mensal - Avaliação de Sazonalidade', fontsize=13, fontweight='bold')
plt.ylabel('Volume Médio (m³)')
plt.xlabel('Mês')
plt.grid(True)
plt.tight_layout()
plt.show()

# Interpretação automática
variacao = df_mes['VOLUME_M3'].max() - df_mes['VOLUME_M3'].min()
coef_var = (variacao / df_mes['VOLUME_M3'].mean()) * 100

print("=== 3️⃣ Análise de sazonalidade ===")
print(f"Coeficiente de variação mensal: {coef_var:.2f}%")

if coef_var > 30:
    print("✅ O comportamento apresenta indícios de SAZONALIDADE — há variação perceptível entre os meses.")
    print("Os picos e vales indicam meses com maior e menor atividade de importação/exportação.\n")
else:
    print("⚪ O comportamento parece ALEATÓRIO — as médias mensais são relativamente estáveis.\n")

print("=== Conclusão Geral ===")
print("1️⃣ Meses no quartil superior: ver listagem acima (maior frequência indica meses mais fortes).")
print("2️⃣ Outliers: presença confirmada pelo critério de 1,5 × IQR.")
print("3️⃣ Sazonalidade: comportamento sazonal se CV > 30%, aleatório se menor.")
```

```python
# =====================================================
# ETAPA 3 – PADRONIZAÇÃO (Z-SCORE) E DISTRIBUIÇÃO NORMAL
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -----------------------------------------------------
# 1️⃣ Leitura do arquivo limpo da Etapa 1
# -----------------------------------------------------
file_path = "importacoes_exportacoes_limpos_etapa1_corrigido.csv"
df = pd.read_csv(file_path)

# Considerar apenas valores válidos de volume
df = df[['VOLUME_M3']].dropna()
print(f"Registros válidos: {len(df)}")

# -----------------------------------------------------
# 2️⃣ Cálculo da média e desvio-padrão
# -----------------------------------------------------
media = df['VOLUME_M3'].mean()
desvio = df['VOLUME_M3'].std()

# -----------------------------------------------------
# 3️⃣ Padronização (Z-Score)
# -----------------------------------------------------
df['Z_SCORE'] = (df['VOLUME_M3'] - media) / desvio

# -----------------------------------------------------
# 4️⃣ Pergunta 1 – Quantos valores estão acima de 1 desvio-padrão?
# -----------------------------------------------------
acima_1dp = (df['Z_SCORE'] > 1).sum()
perc_acima_1dp = (acima_1dp / len(df)) * 100

print("=== 1️⃣ Valores acima de 1 desvio-padrão ===")
print(f"Quantidade: {acima_1dp} registros ({perc_acima_1dp:.2f}% do total)\n")

# -----------------------------------------------------
# 5️⃣ Pergunta 2 – Probabilidade de observar valor acima de Z = 1,96
# -----------------------------------------------------
prob_acima_196 = 1 - stats.norm.cdf(1.96)  # área da cauda à direita
print("=== 2️⃣ Probabilidade de observar valor acima de Z = 1,96 ===")
print(f"Probabilidade teórica (Distribuição Normal Padrão): {prob_acima_196*100:.2f}%\n")

# -----------------------------------------------------
# 6️⃣ Pergunta 3 – O conjunto é normalmente distribuído?
# -----------------------------------------------------
# Amostra para teste (Shapiro-Wilk e D’Agostino-Pearson)
amostra = df['VOLUME_M3'].sample(500, random_state=42)

shapiro_stat, shapiro_p = stats.shapiro(amostra)
dagostino_stat, dagostino_p = stats.normaltest(amostra)

print("=== 3️⃣ Testes de Normalidade ===")
print(f"Shapiro-Wilk p-valor: {shapiro_p:.4f}")
print(f"D’Agostino-Pearson p-valor: {dagostino_p:.4f}")

if shapiro_p > 0.05 and dagostino_p > 0.05:
    normalidade = "Sim"
    print("\n✅ Os p-valores são maiores que 0,05 → não há evidência contra a normalidade.")
    print("O conjunto pode ser considerado aproximadamente normal.\n")
else:
    normalidade = "Não"
    print("\n⚠️ Os p-valores são menores que 0,05 → rejeita-se a hipótese de normalidade.")
    print("A distribuição não é perfeitamente normal (pode haver assimetria ou caudas longas).\n")

# -----------------------------------------------------
# 7️⃣ Visualização gráfica (Histograma + Curva Normal)
# -----------------------------------------------------
plt.figure(figsize=(10,5))
sns.histplot(df['Z_SCORE'], bins=50, kde=True, color='skyblue', stat='density')
x = np.linspace(-4,4,100)
plt.plot(x, stats.norm.pdf(x, 0, 1), color='red', lw=2, label='Curva Normal Padrão')
plt.axvline(1, color='orange', linestyle='--', label='Z = 1')
plt.axvline(1.96, color='green', linestyle='--', label='Z = 1.96')
plt.title('Distribuição dos Z-Scores e Curva Normal Padrão', fontsize=13, fontweight='bold')
plt.xlabel('Z-Score')
plt.ylabel('Densidade')
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------------------------------
# 8️⃣ Conclusão resumida
# -----------------------------------------------------
print("=== CONCLUSÃO FINAL ===")
print(f"1️⃣ {acima_1dp} valores ({perc_acima_1dp:.2f}%) estão acima de 1 desvio-padrão da média.")
print(f"2️⃣ A probabilidade teórica de observar um valor acima de Z = 1,96 é {prob_acima_196*100:.2f}%.")
print(f"3️⃣ O conjunto pode ser considerado normalmente distribuído? {normalidade}.")
```

```python
# =====================================================
# ETAPA 4 – INFERÊNCIA E INTERVALOS DE CONFIANÇA (95%)
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -----------------------------------------------------
# 1️⃣ Leitura dos dados já limpos (Etapa 1)
# -----------------------------------------------------
file_path = "importacoes_exportacoes_limpos_etapa1_corrigido.csv"
df = pd.read_csv(file_path)

# Remove valores nulos
df = df[['VOLUME_M3', 'DISPENDIO_USD_FOB']].dropna()
n_pop = len(df)
print(f"Tamanho populacional: {n_pop} registros")

# -----------------------------------------------------
# 2️⃣ Selecionar uma amostra aleatória de 300 observações
# -----------------------------------------------------
n_amostra = 300
amostra = df.sample(n=n_amostra, random_state=42)
print(f"Amostra aleatória selecionada: {len(amostra)} observações\n")

# -----------------------------------------------------
# 3️⃣ Intervalo de confiança de 95% para a média do Volume
# -----------------------------------------------------
media_amostra_vol = amostra['VOLUME_M3'].mean()
desvio_amostra_vol = amostra['VOLUME_M3'].std(ddof=1)
erro_padrao_vol = desvio_amostra_vol / np.sqrt(n_amostra)

# Valor crítico t para 95% e 299 graus de liberdade
t_critico = stats.t.ppf(0.975, df=n_amostra-1)
margem_erro_vol = t_critico * erro_padrao_vol

ic_inf_vol = media_amostra_vol - margem_erro_vol
ic_sup_vol = media_amostra_vol + margem_erro_vol

print("=== Intervalo de Confiança (95%) - Volume (m³) ===")
print(f"Média amostral: {media_amostra_vol:,.2f}")
print(f"Erro padrão: {erro_padrao_vol:,.2f}")
print(f"Margem de erro (E): {margem_erro_vol:,.2f}")
print(f"IC 95%: [{ic_inf_vol:,.2f} , {ic_sup_vol:,.2f}]\n")

# -----------------------------------------------------
# 4️⃣ Intervalo de confiança de 95% para a média do Dispêndio
# -----------------------------------------------------
media_amostra_disp = amostra['DISPENDIO_USD_FOB'].mean()
desvio_amostra_disp = amostra['DISPENDIO_USD_FOB'].std(ddof=1)
erro_padrao_disp = desvio_amostra_disp / np.sqrt(n_amostra)

margem_erro_disp = t_critico * erro_padrao_disp
ic_inf_disp = media_amostra_disp - margem_erro_disp
ic_sup_disp = media_amostra_disp + margem_erro_disp

print("=== Intervalo de Confiança (95%) - Dispêndio (US$ FOB) ===")
print(f"Média amostral: {media_amostra_disp:,.2f}")
print(f"Erro padrão: {erro_padrao_disp:,.2f}")
print(f"Margem de erro (E): {margem_erro_disp:,.2f}")
print(f"IC 95%: [{ic_inf_disp:,.2f} , {ic_sup_disp:,.2f}]\n")

# -----------------------------------------------------
# 5️⃣ Comparação com as médias populacionais
# -----------------------------------------------------
media_pop_vol = df['VOLUME_M3'].mean()
media_pop_disp = df['DISPENDIO_USD_FOB'].mean()

print("=== Comparação com médias populacionais ===")
print(f"Média populacional (Volume): {media_pop_vol:,.2f}")
print(f"Média populacional (Dispêndio): {media_pop_disp:,.2f}\n")

# Pergunta 1 – O intervalo contém o valor médio populacional?
contido_vol = ic_inf_vol <= media_pop_vol <= ic_sup_vol
contido_disp = ic_inf_disp <= media_pop_disp <= ic_sup_disp

print("=== 1️⃣ O intervalo contém o valor médio populacional? ===")
print(f"Volume: {'✅ Sim' if contido_vol else '❌ Não'}")
print(f"Dispêndio: {'✅ Sim' if contido_disp else '❌ Não'}\n")

# -----------------------------------------------------
# 6️⃣ Visualização dos Intervalos de Confiança
# -----------------------------------------------------
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Volume
ax[0].errorbar(1, media_amostra_vol, yerr=margem_erro_vol, fmt='o', color='teal', capsize=8)
ax[0].axhline(media_pop_vol, color='red', linestyle='--', label='Média Populacional')
ax[0].set_title('IC 95% - Média do Volume (m³)')
ax[0].set_ylabel('Volume (m³)')
ax[0].set_xticks([])
ax[0].legend()

# Dispêndio
ax[1].errorbar(1, media_amostra_disp, yerr=margem_erro_disp, fmt='o', color='orange', capsize=8)
ax[1].axhline(media_pop_disp, color='red', linestyle='--', label='Média Populacional')
ax[1].set_title('IC 95% - Média do Dispêndio (US$ FOB)')
ax[1].set_ylabel('Dispêndio (US$)')
ax[1].set_xticks([])
ax[1].legend()

plt.tight_layout()
plt.show()

# -----------------------------------------------------
# 7️⃣ Pergunta 2 – Como o tamanho da amostra influencia a precisão?
# -----------------------------------------------------
print("=== 2️⃣ Como o tamanho da amostra influencia a precisão do intervalo? ===")
print("Quanto maior o tamanho da amostra (n), menor o erro padrão e a margem de erro (E).")
print("Isso significa que o intervalo de confiança fica mais estreito e a estimativa mais precisa.\n")

# -----------------------------------------------------
# 8️⃣ Pergunta 3 – Qual seria o erro máximo admissível (E)?
# -----------------------------------------------------
print("=== 3️⃣ Erro máximo admissível (E) ===")
print(f"Volume: {margem_erro_vol:,.2f}")
print(f"Dispêndio: {margem_erro_disp:,.2f}")
print("\nO erro máximo admissível (E) é a margem de erro associada ao IC de 95%,")
print("indicando o quanto a média amostral pode se afastar da média verdadeira da população.")
```

```python
# =====================================================
# ETAPA 5 – TESTES DE HIPÓTESES (médias e proporções)
# =====================================================

import pandas as pd
import numpy as np
from scipy import stats

# -----------------------------------------------------
# 1️⃣ Leitura do arquivo limpo (Etapa 1)
# -----------------------------------------------------
file_path = "importacoes_exportacoes_limpos_etapa1_corrigido.csv"
df = pd.read_csv(file_path)

# Filtrar produtos relevantes
df = df[['PRODUTO', 'VOLUME_M3', 'DISPENDIO_USD_FOB']].dropna()

# -----------------------------------------------------
# HIPÓTESE 1: Médias de Gasolina A e Óleo Diesel
# -----------------------------------------------------
print("=== HIPÓTESE 1: Comparação entre médias ===\n")

# Filtrar cada produto
gasolina = df[df['PRODUTO'].str.contains("GASOLINA A", case=False, na=False)]['VOLUME_M3']
diesel = df[df['PRODUTO'].str.contains("DIESEL", case=False, na=False)]['VOLUME_M3']

# Garantir que há dados suficientes
print(f"Registros Gasolina A: {len(gasolina)}")
print(f"Registros Óleo Diesel: {len(diesel)}\n")

# --- 1. Definição das hipóteses ---
print("Hipóteses:")
print("H0: μ_GasolinaA = μ_Diesel  → As médias de volume são iguais.")
print("H1: μ_GasolinaA ≠ μ_Diesel  → As médias de volume são diferentes.\n")

# --- 2. Teste t para amostras independentes ---
t_stat, p_valor = stats.ttest_ind(gasolina, diesel, equal_var=False, nan_policy='omit')

print("Resultado do teste t:")
print(f"t = {t_stat:.4f}")
print(f"p-valor = {p_valor:.6f}")

# --- 3. Interpretação (α = 0,05) ---
alpha = 0.05
if p_valor < alpha:
    print("\n✅ Rejeita-se H0: há diferença estatisticamente significativa entre as médias.")
else:
    print("\n⚪ Não se rejeita H0: não há diferença significativa entre as médias.")

# -----------------------------------------------------
# HIPÓTESE 2: Proporção de operações com dispêndio > US$ 500 mil
# -----------------------------------------------------
print("\n\n=== HIPÓTESE 2: Teste de proporção ===\n")

# Criação da variável binária (1 se dispêndio > 500.000)
df['ACIMA_500K'] = (df['DISPENDIO_USD_FOB'] > 500000).astype(int)

# Parâmetros observados
n = len(df)
x = df['ACIMA_500K'].sum()
p_amostra = x / n
p_hipotese = 0.5

print(f"Número de operações: {n}")
print(f"Operações com dispêndio > 500 mil: {x}")
print(f"Proporção amostral: {p_amostra:.4f}\n")

print("Hipóteses:")
print("H0: p = 0,5  → A proporção é igual a 50%.")
print("H1: p ≠ 0,5  → A proporção difere de 50%.\n")

# --- 2. Teste Z para proporção ---
z_stat = (p_amostra - p_hipotese) / np.sqrt((p_hipotese * (1 - p_hipotese)) / n)
p_valor_prop = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Resultado do teste z:")
print(f"z = {z_stat:.4f}")
print(f"p-valor = {p_valor_prop:.6f}")

# --- 3. Interpretação (α = 0,05) ---
if p_valor_prop < alpha:
    print("\n✅ Rejeita-se H0: a proporção difere significativamente de 0,5.")
else:
    print("\n⚪ Não se rejeita H0: não há diferença significativa em relação a 0,5.")

# -----------------------------------------------------
# Gráficos ilustrativos
# -----------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))

# Gráfico 1 – Médias comparadas
plt.subplot(1,2,1)
sns.barplot(x=['Gasolina A', 'Óleo Diesel'],
            y=[gasolina.mean(), diesel.mean()],
            palette=['skyblue','lightgreen'])
plt.title("Comparação das Médias de Volume (m³)")
plt.ylabel("Média do Volume (m³)")

# Gráfico 2 – Proporção de operações acima de 500 mil
plt.subplot(1,2,2)
sns.barplot(x=['≤ 500 mil', '> 500 mil'],
            y=[1 - p_amostra, p_amostra],
            palette=['lightcoral','teal'])
plt.title("Proporção de Operações com Dispêndio > US$ 500 mil")
plt.ylabel("Proporção Observada")
plt.tight_layout()
plt.show()

# -----------------------------------------------------
# Conclusão geral
# -----------------------------------------------------
print("\n=== CONCLUSÃO FINAL ===")
print("1️⃣ Hipótese 1: Teste t para médias →", "Rejeita H0 (diferença significativa)" if p_valor < alpha else "Não rejeita H0 (sem diferença significativa)")
print("2️⃣ Hipótese 2: Teste z para proporção →", "Rejeita H0 (proporção diferente de 0,5)" if p_valor_prop < alpha else "Não rejeita H0 (sem diferença significativa)")
```
