```python
# separar e padronizar colunas
import pandas as pd
import numpy as np
import re
from pathlib import Path

# Corrigindo nome do arquivo se
input_file = "importacoes-exportacoes-derivados-2000-2025.csv"
output_file = "importacoes_exportacoes_colunas_corrigido.csv"

# 1) Leitura robusta: tenta detectar separador automaticamente
# (usa engine 'python' para permitir sep=None)
try:
    df_raw = pd.read_csv(input_file, sep=None, engine='python', encoding='utf-8')
except Exception as e:
    # fallback: ler como texto e depois splitar por ';'
    txt = open(input_file, 'r', encoding='utf-8', errors='replace').read()
    rows = [r.split(';') for r in txt.splitlines() if r.strip()!='']
    df_raw = pd.DataFrame(rows[1:], columns=rows[0])

print("Colunas lidas inicialmente:", list(df_raw.columns)[:10])

# 2) Se apenas 1 coluna foi lida (arquivo com tudo em uma string), tenta split manual por ';'
if len(df_raw.columns) == 1:
    col0 = df_raw.columns[0]
    # tenta separar por ';' e se falhar, por ','
    sample = df_raw[col0].iloc[:5].astype(str).tolist()
    sep = ';' if any(';' in s for s in sample) else (',' if any(',' in s for s in sample) else None)
    if sep is None:
        # tentar detectar por número de campos com regex (padrão CSV com ;)
        sep = ';'
    df_split = df_raw[col0].str.split(sep, expand=True)
    # assumir primeira linha como cabeçalho se parecer com nomes
    header = df_split.iloc[0].tolist()
    df_split.columns = header
    df_raw = df_split.drop(index=0).reset_index(drop=True)
    print("Split manual aplicado usando separador:", repr(sep))
    print("Novas colunas:", list(df_raw.columns)[:10])

# 3) Normaliza nomes de colunas (remove espaços extras, maiúsculas)
cols_orig = [str(c).strip() for c in df_raw.columns]
cols_up = [c.upper() for c in cols_orig]
df_raw.columns = cols_up

# 4) Mapeamento para os nomes exatos solicitados
rename_map = {}
for c in df_raw.columns:
    c_u = c.upper()
    if 'ANO' in c_u:
        rename_map[c] = 'Ano'
    elif 'MÊS' in c_u or 'MES' in c_u:
        rename_map[c] = 'Mês'
    elif 'PRODUT' in c_u or 'PROD' in c_u:
        rename_map[c] = 'Produto'
    elif 'OPERA' in c_u or 'OPERAÇÃO' in c_u or 'OPERACAO' in c_u:
        rename_map[c] = 'Operação Comercial'
    elif ('IMPORT' in c_u and 'EXPORT' not in c_u) or ('EXPORT' in c_u and 'IMPORT' not in c_u) or 'IMPORTADO' in c_u or 'EXPORTADO' in c_u or 'M3' in c_u or 'M³' in c_u or 'VOLUME' in c_u:
        # pode ser "IMPORTADO / EXPORTADO" ou similar
        rename_map[c] = 'Importado/Exportado (m³)'
    elif 'DISP' in c_u or 'RECEITA' in c_u or 'US$' in c_u or 'FOB' in c_u:
        rename_map[c] = 'Dispêndio/Receita (US$ FOB)'
    else:
        # deixa o original formatado minimamente
        rename_map[c] = c.title()

df = df_raw.rename(columns=rename_map)

# 5) Seleciona apenas as colunas desejadas (mantém somente as que existirem)
colunas_desejadas = [
    'Ano',
    'Mês',
    'Produto',
    'Operação Comercial',
    'Importado/Exportado (m³)',
    'Dispêndio/Receita (US$ FOB)'
]
colunas_encontradas = [c for c in colunas_desejadas if c in df.columns]
if len(colunas_encontradas) < len(colunas_desejadas):
    missing = [c for c in colunas_desejadas if c not in df.columns]
    print("Atenção: colunas esperadas não encontradas no arquivo original:", missing)

df = df[colunas_encontradas].copy()

# 6) Limpeza das colunas numéricas: tratar '.' e ',' e remover caracteres não numéricos
def to_numeric_col(s):
    # aceita strings com "." como milhar e "," como decimal ou vice-versa
    def conv(x):
        if pd.isna(x): return np.nan
        xs = str(x).strip()
        if xs == '': return np.nan
        # remover espaços
        xs = xs.replace(' ', '')
        # se tiver both '.' and ',' assume '.' milhar e ',' decimal -> remove '.' e replace ','->'.'
        if xs.count('.') > 0 and xs.count(',') > 0:
            xs = xs.replace('.', '').replace(',', '.')
        else:
            # se tiver apenas '.', pode ser decimal ou milhar; tentar interpretar:
            # se '.' aparece mais de once (unlikely) or length>3 and last part length==3 assume milhar
            if xs.count('.') > 0 and len(xs.split('.')[-1]) == 3:
                xs = xs.replace('.', '')
            # se tiver apenas ',' assume decimal
            if xs.count(',') > 0 and xs.count('.') == 0:
                xs = xs.replace('.', '').replace(',', '.')
        # remover qualquer caractere não numérico exceto minus e dot
        xs = re.sub(r'[^\d\.-]', '', xs)
        try:
            return float(xs)
        except:
            return np.nan
    return s.apply(conv)

# Aplica conversão nas colunas encontradas
if 'Importado/Exportado (m³)' in df.columns:
    df['Importado/Exportado (m³)'] = to_numeric_col(df['Importado/Exportado (m³)'])
if 'Dispêndio/Receita (US$ FOB)' in df.columns:
    df['Dispêndio/Receita (US$ FOB)'] = to_numeric_col(df['Dispêndio/Receita (US$ FOB)'])

# 7) Convertendo Ano para inteiro quando possível; padronizando Mês (capitalizado)
if 'Ano' in df.columns:
    df['Ano'] = pd.to_numeric(df['Ano'], errors='coerce').astype('Int64')
if 'Mês' in df.columns:
    df['Mês'] = df['Mês'].astype(str).str.strip().str.title()

# 8) Remover linhas completamente vazias e duplicatas
df = df.dropna(how='all').drop_duplicates().reset_index(drop=True)

# 9) (Opcional) Tratar valores inválidos: considerar <=0 como NaN para análise
if 'Importado/Exportado (m³)' in df.columns:
    df.loc[df['Importado/Exportado (m³)'] <= 0, 'Importado/Exportado (m³)'] = np.nan
if 'Dispêndio/Receita (US$ FOB)' in df.columns:
    df.loc[df['Dispêndio/Receita (US$ FOB)'] <= 0, 'Dispêndio/Receita (US$ FOB)'] = np.nan

# 10) Salvar CSV corrigido (com encoding compatível)
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"\n✅ Arquivo salvo: {output_file}")

# 11) Mostrar resumo e amostra
print("\nResumo após padronização:")
print("Colunas finais:", df.columns.tolist())
print("Registros totais:", len(df))
if 'Ano' in df.columns:
    anos = df['Ano'].dropna().astype(int)
    if len(anos)>0:
        print(f"Período temporal: {int(anos.min())} a {int(anos.max())}")
print("\nContagem de valores não-nulos por coluna:")
print(df.notna().sum())

print("\nAmostra (primeiras 10 linhas):")
display(df.head(10))
```

```python
# ============================================
# ETAPA 1 – LIMPEZA E ORGANIZAÇÃO DOS DADOS
# ============================================

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------------
# 1️⃣ Leitura do arquivo
# -------------------------------------------------
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path)

print("Colunas encontradas no CSV:")
print(list(df.columns))
print(f"\nTotal de registros originais: {len(df)}")

# -------------------------------------------------
# 2️⃣ Padroniza nomes de colunas (maiúsculas e sem acentos)
# -------------------------------------------------
df.columns = [c.strip().upper().replace('Ã','A').replace('Õ','O').replace('Ç','C').replace('É','E').replace('Ê','E').replace('Í','I').replace('Ú','U').replace('Ó','O').replace('Â','A') for c in df.columns]

# Identificação automática das colunas principais
col_ano = next((c for c in df.columns if 'ANO' in c), None)
col_mes = next((c for c in df.columns if 'MES' in c or 'MÊS' in c), None)
col_prod = next((c for c in df.columns if 'PRODUT' in c), None)
col_op = next((c for c in df.columns if 'OPERACAO' in c or 'OPERACAO COMERCIAL' in c or 'TIPO' in c), None)
col_vol = next((c for c in df.columns if 'M3' in c or 'M³' in c or 'VOLUME' in c or 'QUANT' in c or 'QTDE' in c or 'IMPORTADO' in c or 'EXPORTADO' in c), None)
col_disp = next((c for c in df.columns if 'DISP' in c or 'FOB' in c or 'US$' in c or 'RECEITA' in c or 'VALOR' in c), None)

print("\nMapeamento automático detectado:")
print(f"Ano → {col_ano}")
print(f"Mês → {col_mes}")
print(f"Produto → {col_prod}")
print(f"Operação Comercial → {col_op}")
print(f"Volume → {col_vol}")
print(f"Dispêndio → {col_disp}")

# -------------------------------------------------
# 3️⃣ Seleciona e renomeia colunas
# -------------------------------------------------
rename_map = {
    col_ano: 'Ano',
    col_mes: 'Mês',
    col_prod: 'Produto',
    col_op: 'Operação Comercial',
    col_vol: 'Volume (m³)',
    col_disp: 'Dispêndio (US$ FOB)'
}
df = df.rename(columns=rename_map)
df = df[list(rename_map.values())].copy()

# -------------------------------------------------
# 4️⃣ Conversões numéricas e limpeza
# -------------------------------------------------
def to_numeric_safe(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str):
        x2 = x.replace('.', '').replace(',', '.')
        x2 = re.sub(r'[^\d\.-]', '', x2)
        try:
            return float(x2)
        except:
            return np.nan
    return float(x)

df['Volume (m³)'] = df['Volume (m³)'].apply(to_numeric_safe)
df['Dispêndio (US$ FOB)'] = df['Dispêndio (US$ FOB)'].apply(to_numeric_safe)

# Remove duplicatas e valores inválidos
df = df.drop_duplicates()
df.loc[df['Volume (m³)'] <= 0, 'Volume (m³)'] = np.nan
df.loc[df['Dispêndio (US$ FOB)'] <= 0, 'Dispêndio (US$ FOB)'] = np.nan

# -------------------------------------------------
# 5️⃣ Respostas às perguntas
# -------------------------------------------------
print("\n=== RESULTADOS ETAPA 1 ===")

# 1️⃣ Quantos registros válidos ficaram após limpeza
registros_validos = len(df.dropna(subset=['Volume (m³)', 'Dispêndio (US$ FOB)'], how='all'))
print(f"1️⃣ Registros válidos após limpeza: {registros_validos}")

# 2️⃣ Período temporal coberto
anos_validos = df['Ano'].dropna().astype(int)
ano_min, ano_max = anos_validos.min(), anos_validos.max()
print(f"2️⃣ Período temporal: {ano_min} a {ano_max}")

# 3️⃣ Outliers (média ± 3×desvio padrão)
def detectar_outliers(col):
    serie = df[col].dropna()
    media = serie.mean()
    desvio = serie.std()
    lim_inf = media - 3 * desvio
    lim_sup = media + 3 * desvio
    outliers = serie[(serie < lim_inf) | (serie > lim_sup)]
    return media, desvio, len(outliers), lim_inf, lim_sup

media_v, desvio_v, out_v, inf_v, sup_v = detectar_outliers('Volume (m³)')
media_d, desvio_d, out_d, inf_d, sup_d = detectar_outliers('Dispêndio (US$ FOB)')

print("\n3️⃣ Outliers detectados:")
print(f"• Volume: média={media_v:,.2f}, desvio={desvio_v:,.2f}, outliers={out_v}")
print(f"• Dispêndio: média={media_d:,.2f}, desvio={desvio_d:,.2f}, outliers={out_d}")

# -------------------------------------------------
# 6️⃣ Visualização – Boxplots
# -------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.boxplot(y=df['Volume (m³)'], color='skyblue', ax=axes[0])
axes[0].set_title("Boxplot – Volume (m³)")
sns.boxplot(y=df['Dispêndio (US$ FOB)'], color='lightgreen', ax=axes[1])
axes[1].set_title("Boxplot – Dispêndio (US$ FOB)")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 7️⃣ Conclusão automática
# -------------------------------------------------
print("\n=== CONCLUSÃO ETAPA 1 ===")
print(f"Após a limpeza, restaram {registros_validos} registros válidos.")
print(f"O período temporal cobre {ano_min} a {ano_max}.")
print(f"Foram detectados {out_v} outliers em volume e {out_d} em dispêndio "
      "(critério: média ± 3×desvio padrão).")
```

```python
# ============================================
# ETAPA 2 – Quartis, Percentis e Outliers
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------------
# 1️⃣ Leitura dos dados limpos da Etapa 1
# -------------------------------------------------
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path)

print(f"Registros carregados: {len(df)}")
print(f"Colunas disponíveis: {list(df.columns)}")

# **FIX:** Convert 'Importado/Exportado (m³)' to numeric after loading
if 'Importado/Exportado (m³)' in df.columns:
    df['Importado/Exportado (m³)'] = pd.to_numeric(
        df['Importado/Exportado (m³)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    # Also apply the > 0 filter as done in the previous cell for consistency
    df.loc[df['Importado/Exportado (m³)'] <= 0, 'Importado/Exportado (m³)'] = np.nan

# -------------------------------------------------
# 2️⃣ Seleção da variável principal (Volume)
# -------------------------------------------------
# Remove valores nulos
df = df.dropna(subset=['Importado/Exportado (m³)']).copy()

# -------------------------------------------------
# 3️⃣ Cálculo de quartis, amplitude e percentis
# -------------------------------------------------
q1 = df['Importado/Exportado (m³)'].quantile(0.25)
q2 = df['Importado/Exportado (m³)'].quantile(0.50)
q3 = df['Importado/Exportado (m³)'].quantile(0.75)
amplitude = q3 - q1

p10 = df['Importado/Exportado (m³)'].quantile(0.10)
p25 = df['Importado/Exportado (m³)'].quantile(0.25)
p75 = df['Importado/Exportado (m³)'].quantile(0.75)
p90 = df['Importado/Exportado (m³)'].quantile(0.90)

print("\n=== Quartis e Percentis ===")
print(f"Q1 (25%) = {q1:,.2f}")
print(f"Q2 (Mediana, 50%) = {q2:,.2f}")
print(f"Q3 (75%) = {q3:,.2f}")
print(f"Amplitude Interquartil (A = Q3 - Q1) = {amplitude:,.2f}")
print(f"P10 = {p10:,.2f} | P25 = {p25:,.2f} | P75 = {p75:,.2f} | P90 = {p90:,.2f}")

# -------------------------------------------------
# 4️⃣ Identificação do quartil superior (Q3)
# -------------------------------------------------
quartil_superior = df[df['Importado/Exportado (m³)'] > q3]
meses_quartil_superior = (
    quartil_superior['Mês']
    .value_counts()
    .sort_values(ascending=False)
)

print("\n1️⃣ Meses que representam o quartil superior da série (acima de Q3):")
print(meses_quartil_superior)

# -------------------------------------------------
# 5️⃣ Verificação de outliers (Critério 1,5 × A)
# -------------------------------------------------
limite_inferior = q1 - 1.5 * amplitude
limite_superior = q3 + 1.5 * amplitude
outliers = df[(df['Importado/Exportado (m³)'] < limite_inferior) | (df['Importado/Exportado (m³)'] > limite_superior)]

print("\n2️⃣ Há evidência de outliers extremos (1,5 × A)?")
if len(outliers) > 0:
    print(f"✅ Sim, {len(outliers)} valores estão fora dos limites (outliers extremos detectados).")
else:
    print("❎ Não foram detectados outliers extremos.")

# -------------------------------------------------
# 6️⃣ Gráficos – Boxplot e Histograma
# -------------------------------------------------
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y=df['Importado/Exportado (m³)'], color='lightblue')
plt.title("Boxplot – Detecção de Outliers (Volume m³)")
plt.ylabel("Volume (m³)")

plt.subplot(1,2,2)
sns.histplot(df['Importado/Exportado (m³)'], bins=50, kde=True, color='lightgreen')
plt.title("Histograma – Distribuição dos Volumes")
plt.xlabel("Volume (m³)")

plt.tight_layout()
plt.show()

# -------------------------------------------------
# 7️⃣ Sazonalidade – média de volume por mês
# -------------------------------------------------
df_mes = df.groupby('Mês', as_index=False)['Importado/Exportado (m³)'].mean()

# Para manter a ordem cronológica dos meses
ordem_meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho',
               'Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']
df_mes['Mês'] = pd.Categorical(df_mes['Mês'], categories=ordem_meses, ordered=True)
df_mes = df_mes.sort_values('Mês')

plt.figure(figsize=(10,4))
sns.lineplot(data=df_mes, x='Mês', y='Importado/Exportado (m³)', marker='o', color='teal', linewidth=2)
plt.title('Volume Médio Mensal – Análise de Sazonalidade', fontsize=13, fontweight='bold')
plt.ylabel('Volume Médio (m³)')
plt.xlabel('Mês')
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 8️⃣ Cálculo e Interpretação da Sazonalidade
# -------------------------------------------------
variacao = df_mes['Importado/Exportado (m³)'].max() - df_mes['Importado/Exportado (m³)'].min()
coef_var = (variacao / df_mes['Importado/Exportado (m³)'].mean()) * 100

print("\n3️⃣ O comportamento é sazonal ou aleatório?")
print(f"Coeficiente de variação mensal: {coef_var:.2f}%")

if coef_var > 30:
    print("✅ O comportamento é SAZONAL – há variação significativa entre os meses.")
else:
    print("⚪ O comportamento é ALEATÓRIO – os volumes mensais são relativamente estáveis.")

# -------------------------------------------------
# 9️⃣ Conclusão resumida da Etapa 2
# -------------------------------------------------
print("\n=== CONCLUSÃO ETAPA 2 ===")
print(f"• Quartil superior: meses com maiores volumes → {', '.join(meses_quartil_superior.index[:3])}.")
print(f"• Outliers extremos: {'Sim' if len(outliers)>0 else 'Não'}.")
print(f"• Sazonalidade: {'Sazonal' if coef_var>30 else 'Aleatória'}.")
```

```python
# ============================================
# ETAPA 3 – Distribuição Normal e Z-Score
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -------------------------------------------------
# 1️⃣ Leitura dos dados limpos
# -------------------------------------------------
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path)

print(f"Registros carregados: {len(df)}")
print(f"Colunas disponíveis: {list(df.columns)}")

# **FIX:** Convert 'Importado/Exportado (m³)' to numeric after loading
if 'Importado/Exportado (m³)' in df.columns:
    df['Importado/Exportado (m³)'] = pd.to_numeric(
        df['Importado/Exportado (m³)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    # Also apply the > 0 filter for consistency
    df.loc[df['Importado/Exportado (m³)'] <= 0, 'Importado/Exportado (m³)'] = np.nan

# -------------------------------------------------
# 2️⃣ Considerar apenas os valores válidos de Volume
# -------------------------------------------------
df = df.dropna(subset=['Importado/Exportado (m³)']).copy()
vol = df['Importado/Exportado (m³)']

# -------------------------------------------------
# 3️⃣ Cálculo da média e desvio-padrão
# -------------------------------------------------
media = vol.mean()
desvio = vol.std()
print(f"\nMédia do Volume: {media:,.2f}")
print(f"Desvio-padrão: {desvio:,.2f}")

# -------------------------------------------------
# 4️⃣ Padronização – Cálculo do Z-Score
# -------------------------------------------------
df['Z-Score'] = (vol - media) / desvio

# -------------------------------------------------
# 5️⃣ Pergunta 1 – Quantos valores estão acima de 1 desvio-padrão?
# -------------------------------------------------
acima_1dp = (df['Z-Score'] > 1).sum()
perc_acima_1dp = (acima_1dp / len(df)) * 100
print("\n1️⃣ Quantos valores estão acima de 1 desvio-padrão da média?")
print(f"→ {acima_1dp} registros ({perc_acima_1dp:.2f}% do total)")

# -------------------------------------------------
# 6️⃣ Pergunta 2 – Probabilidade de observar Z > 1,96
# -------------------------------------------------
prob_acima_196 = 1 - stats.norm.cdf(1.96)
print("\n2️⃣ Probabilidade de observar um valor acima de Z = 1,96:")
print(f"→ Probabilidade teórica: {prob_acima_196*100:.2f}%")

# -------------------------------------------------
# 7️⃣ Pergunta 3 – O conjunto pode ser considerado normalmente distribuído?
# -------------------------------------------------
# Teste de Shapiro-Wilk e D’Agostino-Pearson
amostra = vol.sample(500, random_state=42) if len(vol) > 500 else vol

shapiro_stat, shapiro_p = stats.shapiro(amostra)
dagostino_stat, dagostino_p = stats.normaltest(amostra)

print("\n3️⃣ Testes de Normalidade:")
print(f"Shapiro-Wilk p-valor = {shapiro_p:.4f}")
print(f"D’Agostino-Pearson p-valor = {dagostino_p:.4f}")

if shapiro_p > 0.05 and dagostino_p > 0.05:
    normalidade = "Sim"
    justificativa = "Os p-valores são maiores que 0,05 → não há evidência contra a normalidade."
else:
    normalidade = "Não"
    justificativa = "Os p-valores são menores que 0,05 → rejeita-se a hipótese de normalidade (distribuição assimétrica)."

print(f"\nO conjunto pode ser considerado normalmente distribuído? {normalidade}.")
print("Justificativa:", justificativa)

# -------------------------------------------------
# 8️⃣ Gráfico – Distribuição Normal e Z-Scores
# -------------------------------------------------
plt.figure(figsize=(10,5))
sns.histplot(df['Z-Score'], bins=50, kde=True, stat='density', color='skyblue')
x = np.linspace(-4, 4, 100)
plt.plot(x, stats.norm.pdf(x, 0, 1), 'r-', lw=2, label='Curva Normal Padrão')
plt.axvline(1, color='orange', linestyle='--', label='Z = 1')
plt.axvline(1.96, color='green', linestyle='--', label='Z = 1.96')
plt.title('Distribuição dos Z-Scores e Curva Normal Padrão')
plt.xlabel('Z-Score')
plt.ylabel('Densidade')
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 9️⃣ Conclusão automática
# -------------------------------------------------
print("\n=== CONCLUSÃO ETAPA 3 ===")
print(f"1️⃣ {acima_1dp} valores ({perc_acima_1dp:.2f}%) estão acima de 1 desvio-padrão.")
print(f"2️⃣ A probabilidade teórica de Z > 1,96 é {prob_acima_196*100:.2f}%")
print(f"3️⃣ Normalidade: {normalidade} ({justificativa})")
```

```python
# ============================================
# ETAPA 4 – Inferência e Intervalos de Confiança (95%)
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -------------------------------------------------
# 1️⃣ Leitura dos dados limpos
# -------------------------------------------------
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path)

# **FIX:** Convert 'Importado/Exportado (m³)' and 'Dispêndio/Receita (US$ FOB)' to numeric after loading
if 'Importado/Exportado (m³)' in df.columns:
    df['Importado/Exportado (m³)'] = pd.to_numeric(
        df['Importado/Exportado (m³)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    df.loc[df['Importado/Exportado (m³)'] <= 0, 'Importado/Exportado (m³)'] = np.nan

if 'Dispêndio/Receita (US$ FOB)' in df.columns:
    df['Dispêndio/Receita (US$ FOB)'] = pd.to_numeric(
        df['Dispêndio/Receita (US$ FOB)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    df.loc[df['Dispêndio/Receita (US$ FOB)'] <= 0, 'Dispêndio/Receita (US$ FOB)'] = np.nan

# Remove valores nulos nas colunas principais (agora usando os nomes corretos)
df = df[['Importado/Exportado (m³)', 'Dispêndio/Receita (US$ FOB)']].dropna()
n_pop = len(df)
print(f"Tamanho populacional: {n_pop} registros válidos\n")

# -------------------------------------------------
# 2️⃣ Seleção de uma amostra aleatória de 300 observações
# -------------------------------------------------
n_amostra = 300
amostra = df.sample(n=n_amostra, random_state=42)
print(f"Amostra selecionada com {n_amostra} observações\n")

# -------------------------------------------------
# 3️⃣ Intervalo de Confiança (95%) – Média do Volume
# -------------------------------------------------
media_amostra_vol = amostra['Importado/Exportado (m³)'].mean()
desvio_amostra_vol = amostra['Importado/Exportado (m³)'].std(ddof=1)
erro_padrao_vol = desvio_amostra_vol / np.sqrt(n_amostra)

# Valor crítico t (95%, gl = n-1)
t_critico = stats.t.ppf(0.975, df=n_amostra-1)

# Margem de erro
margem_erro_vol = t_critico * erro_padrao_vol

# Limites do intervalo
ic_inf_vol = media_amostra_vol - margem_erro_vol
ic_sup_vol = media_amostra_vol + margem_erro_vol

print("=== Intervalo de Confiança (95%) – Volume (m³) ===")
print(f"Média amostral: {media_amostra_vol:,.2f}")
print(f"Erro padrão: {erro_padrao_vol:,.2f}")
print(f"Margem de erro (E): {margem_erro_vol:,.2f}")
print(f"IC95%: [{ic_inf_vol:,.2f} , {ic_sup_vol:,.2f}]\n")

# -------------------------------------------------
# 4️⃣ Intervalo de Confiança (95%) – Média do Dispêndio
# -------------------------------------------------
media_amostra_disp = amostra['Dispêndio/Receita (US$ FOB)'].mean()
desvio_amostra_disp = amostra['Dispêndio/Receita (US$ FOB)'].std(ddof=1)
erro_padrao_disp = desvio_amostra_disp / np.sqrt(n_amostra)
margem_erro_disp = t_critico * erro_padrao_disp

ic_inf_disp = media_amostra_disp - margem_erro_disp
ic_sup_disp = media_amostra_disp + margem_erro_disp

print("=== Intervalo de Confiança (95%) – Dispêndio (US$ FOB) ===")
print(f"Média amostral: {media_amostra_disp:,.2f}")
print(f"Erro padrão: {erro_padrao_disp:,.2f}")
print(f"Margem de erro (E): {margem_erro_disp:,.2f}")
print(f"IC95%: [{ic_inf_disp:,.2f} , {ic_sup_disp:,.2f}]\n")

# -------------------------------------------------
# 5️⃣ Comparação com as médias populacionais
# -------------------------------------------------
media_pop_vol = df['Importado/Exportado (m³)'].mean()
media_pop_disp = df['Dispêndio/Receita (US$ FOB)'].mean()

print("=== Comparação com as Médias Populacionais ===")
print(f"Média populacional (Volume): {media_pop_vol:,.2f}")
print(f"Média populacional (Dispêndio): {media_pop_disp:,.2f}\n")

# Pergunta 1 – O intervalo contém o valor médio populacional?
contido_vol = ic_inf_vol <= media_pop_vol <= ic_sup_vol
contido_disp = ic_inf_disp <= media_pop_disp <= ic_sup_disp

print("1️⃣ O intervalo contém o valor médio populacional?")
print(f"→ Volume: {'✅ Sim' if contido_vol else '❌ Não'}")
print(f"→ Dispêndio: {'✅ Sim' if contido_disp else '❌ Não'}\n")

# -------------------------------------------------
# 6️⃣ Gráficos dos Intervalos de Confiança
# -------------------------------------------------
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Volume
ax[0].errorbar(1, media_amostra_vol, yerr=margem_erro_vol, fmt='o', color='teal', capsize=8)
ax[0].axhline(media_pop_vol, color='red', linestyle='--', label='Média Populacional')
ax[0].set_title('IC95% – Média do Volume (m³)')
ax[0].set_ylabel('Volume (m³)')
ax[0].set_xticks([])
ax[0].legend()

# Dispêndio
ax[1].errorbar(1, media_amostra_disp, yerr=margem_erro_disp, fmt='o', color='orange', capsize=8)
ax[1].axhline(media_pop_disp, color='red', linestyle='--', label='Média Populacional')
ax[1].set_title('IC95% – Média do Dispêndio (US$ FOB)')
ax[1].set_ylabel('Dispêndio (US$)')
ax[1].set_xticks([])
ax[1].legend()

plt.tight_layout()
plt.show()

# -------------------------------------------------
# 7️⃣ Pergunta 2 – Como o tamanho da amostra influencia a precisão?
# -------------------------------------------------
print("2️⃣ Como o tamanho da amostra influencia a precisão?")
print("→ Quanto maior o tamanho da amostra (n), menor o erro padrão e a margem de erro (E).")
print("   Isso significa que o intervalo de confiança fica mais estreito e a estimativa mais precisa.\n")

# -------------------------------------------------
# 8️⃣ Pergunta 3 – Qual é o erro máximo admissível (E)?
# -------------------------------------------------
print("3️⃣ Erro máximo admissível (E):")
print(f"→ Volume: {margem_erro_vol:,.2f}")
print(f"→ Dispêndio: {margem_erro_disp:,.2f}")
print("\nO erro máximo admissível representa o limite máximo de variação esperado entre a média amostral e a média real da população, com 95% de confiança.")

# -------------------------------------------------
# 9️⃣ Conclusão automática da Etapa 4
# -------------------------------------------------
print("\n=== CONCLUSÃO ETAPA 4 ===")
print(f"• IC95% (Volume): [{ic_inf_vol:,.2f} , {ic_sup_vol:,.2f}] → Contém média populacional? {'Sim' if contido_vol else 'Não'}")
print(f"• IC95% (Dispêndio): [{ic_inf_disp:,.2f} , {ic_sup_disp:,.2f}] → Contém média populacional? {'Sim' if contido_disp else 'Não'}")
print(f"• O tamanho da amostra de {n_amostra} registros define o nível de precisão do intervalo.")
print(f"• Erro máximo admissível (Volume: {margem_erro_vol:,.2f}, Dispêndio: {margem_erro_disp:,.2f}).")
```

```python
# =====================================================
# ETAPA 5 – Testes de Hipóteses (Médias e Proporções)
# =====================================================

import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------------------------
# 1️⃣ Leitura do arquivo limpo (Etapa 1)
# -----------------------------------------------------
from google.colab import files
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
df = pd.read_csv(file_path)

# **FIX:** Convert 'Importado/Exportado (m³)' and 'Dispêndio/Receita (US$ FOB)' to numeric after loading
if 'Importado/Exportado (m³)' in df.columns:
    df['Importado/Exportado (m³)'] = pd.to_numeric(
        df['Importado/Exportado (m³)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    df.loc[df['Importado/Exportado (m³)'] <= 0, 'Importado/Exportado (m³)'] = np.nan

if 'Dispêndio/Receita (US$ FOB)' in df.columns:
    df['Dispêndio/Receita (US$ FOB)'] = pd.to_numeric(
        df['Dispêndio/Receita (US$ FOB)'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False),
        errors='coerce'
    )
    df.loc[df['Dispêndio/Receita (US$ FOB)'] <= 0, 'Dispêndio/Receita (US$ FOB)'] = np.nan

# Selecionar apenas colunas relevantes e remover nulos
df = df[['Produto', 'Importado/Exportado (m³)', 'Dispêndio/Receita (US$ FOB)']].dropna()

# -----------------------------------------------------
# HIPÓTESE 1: Médias de Gasolina A e Óleo Diesel
# -----------------------------------------------------
print("=== HIPÓTESE 1: Comparação entre médias de Volume (m³) ===\n")

# Filtrar produtos
gasolina = df[df['Produto'].str.contains("GASOLINA A", case=False, na=False)]['Importado/Exportado (m³)']
diesel = df[df['Produto'].str.contains("DIESEL", case=False, na=False)]['Importado/Exportado (m³)']

print(f"Registros Gasolina A: {len(gasolina)}")
print(f"Registros Óleo Diesel: {len(diesel)}\n")

# 1️⃣ Definição das Hipóteses
print("Hipóteses:")
print("H₀: μ_GasolinaA = μ_Diesel → As médias de volume são iguais.")
print("H₁: μ_GasolinaA ≠ μ_Diesel → As médias de volume são diferentes.\n")

# 2️⃣ Teste t para amostras independentes
t_stat, p_valor = stats.ttest_ind(gasolina, diesel, equal_var=False, nan_policy='omit')

print(f"Estatística t = {t_stat:.4f}")
print(f"p-valor = {p_valor:.6f}")

# 3️⃣ Interpretação com α = 0,05
alpha = 0.05
if p_valor < alpha:
    print("\n✅ Rejeita-se H₀: há diferença estatisticamente significativa entre as médias.\n")
else:
    print("\n⚪ Não se rejeita H₀: não há diferença significativa entre as médias.\n")

# Gráfico ilustrativo
plt.figure(figsize=(6,5))
sns.barplot(x=['Gasolina A', 'Óleo Diesel'], y=[gasolina.mean(), diesel.mean()],
            palette=['skyblue','lightgreen'])
plt.title("Comparação das Médias de Volume (m³)")
plt.ylabel("Volume Médio (m³)")
plt.show()

# -----------------------------------------------------
# HIPÓTESE 2: Proporção de operações com dispêndio > 500 mil
# -----------------------------------------------------
print("\n=== HIPÓTESE 2: Teste de Proporção ===\n")

# Criação da variável binária (1 se dispêndio > 500.000)
df['ACIMA_500K'] = (df['Dispêndio/Receita (US$ FOB)'] > 500000).astype(int)

# Parâmetros observados
n = len(df)
x = df['ACIMA_500K'].sum()
p_amostra = x / n
p_hipotese = 0.5

print(f"Número de operações: {n}")
print(f"Operações acima de 500 mil: {x}")
print(f"Proporção amostral: {p_amostra:.4f}\n")

# 1️⃣ Definição das Hipóteses
print("Hipóteses:")
print("H₀: p = 0,5 → A proporção é igual a 50%.")
print("H₁: p ≠ 0,5 → A proporção difere de 50%.\n")

# 2️⃣ Teste Z para proporção
z_stat = (p_amostra - p_hipotese) / np.sqrt((p_hipotese * (1 - p_hipotese)) / n)
p_valor_prop = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Estatística z = {z_stat:.4f}")
print(f"p-valor = {p_valor_prop:.6f}")

# 3️⃣ Interpretação
if p_valor_prop < alpha:
    print("\n✅ Rejeita-se H₀: a proporção difere significativamente de 0,5.\n")
else:
    print("\n⚪ Não se rejeita H₀: não há diferença significativa em relação a 0,5.\n")

# Gráfico da proporção observada
plt.figure(figsize=(6,5))
sns.barplot(x=['≤ 500 mil', '> 500 mil'],
            y=[1 - p_amostra, p_amostra],
            palette=['lightcoral','teal'])
plt.title("Proporção de Operações com Dispêndio > US$ 500 mil")
plt.ylabel("Proporção Observada")
plt.show()

# -----------------------------------------------------
# Conclusão geral da Etapa 5
# -----------------------------------------------------
print("=== CONCLUSÃO FINAL ===")
print(f"1️⃣ Hipótese 1 (Teste t): {'Rejeita H₀ → há diferença significativa' if p_valor < alpha else 'Não rejeita H₀ → médias equivalentes'}")
print(f"2️⃣ Hipótese 2 (Teste z): {'Rejeita H₀ → proporção diferente de 0,5' if p_valor_prop < alpha else 'Não rejeita H₀ → proporção semelhante a 0,5'}")
```
